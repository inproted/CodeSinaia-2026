This is a demo notebook, showcasing the structure of a `.ipynb` file as a sequence of text and code.  
The following code defines a variable with a random numerical value in between 10 (inclusive) and 90 (exclusive) and prints it in the page.

```python
import random
number = random.randint(10,90)
print(number)
```

Every block of code in a page is contextual to the blocks preceding it in the same document. The block may fail if the context is not correct. The following block increments the number set in previous block, but fails to do so on a variable that was not defined in this page. In case of an error, the error is printed in the same page, under the block.

```python
number += 1
print(number)
# other += 2
# print(other)
```

The execution may include graphical elements (i.e. charts), which are inserted in the page as the execution of the block completes.

```python
import random
import matplotlib.pyplot as plt

random_values = [random.randint(-1000, 1000) for _ in range(100)]

plt.figure(figsize=(6, 2))
plt.boxplot(random_values, vert=False, whis=1.5)
plt.title("Box-and-Whisker Plot for 500 Random Values")
plt.xlabel("Value")
plt.tight_layout()
plt.show()
```
